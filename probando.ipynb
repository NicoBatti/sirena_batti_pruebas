```python
import pandas as pd
import pyodbc
import os

# Ruta al archivo MDB
mdb_file = r"D:\COMPU\Documents\Descargas\AR-SOTER\SOTER_AR\Database\AR_SOTERv1.mdb"
output_folder = r"D:\COMPU\Documents\Descargas"

# Conexión al archivo MDB (Access)
conn_str = (
    r"DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};"
    rf"DBQ={mdb_file};"
)
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()

# Obtener nombres de tablas
tables = [table.table_name for table in cursor.tables(tableType='TABLE')]

# Exportar cada tabla a CSV
for table in tables:
    df = pd.read_sql(f"SELECT * FROM [{table}]", conn)
    csv_path = os.path.join(output_folder, f"{table}.csv")
    df.to_csv(csv_path, index=False, encoding='utf-8')
    print(f"Exportada: {csv_path}")

conn.close()
print("✅ Conversión completa")
```
