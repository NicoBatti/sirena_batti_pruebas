```python
import pandas as pd

# --- 1. Cargar CSV gigante ---
df = pd.read_csv("smap_2025-11-16.csv")

# --- 2. Columnas que queremos conservar ---

cols_keep = [
    # Ubicación y tiempo
    "cell_lat", "cell_lon", "time", "grid_id",

    # Humedad del suelo
    "sm_surface",
    "sm_rootzone",
    "sm_surface_wetness",
    "sm_rootzone_wetness",
    "sm_rootzone_pctl",

    # Escorrentía
    "overland_runoff_flux",
    "baseflow_flux",
    "snow_melt_flux",
    "snowfall_surface_flux",
    "soil_water_infiltration_flux",

    # Evaporación
    "land_evapotranspiration_flux",
    "heat_flux_latent",
]

# --- 3. Filtrar solo esas columnas ---
df_clean = df[cols_keep]

# --- 4. Guardar CSV limpio ---
df_clean.to_csv("smap_clean.csv", index=False)

print("Listo! CSV reducido guardado como smap_clean.csv")
print("Columnas finales:", df_clean.columns.tolist())
print("Tamaño final:", df_clean.memory_usage(deep=True).sum() / 1e6, "MB")
```
